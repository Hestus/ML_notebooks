```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
# Simple example

x = np.arange(0, 10)
y = np.arange(11,21)
```

```python
# Scatter plot

plt.scatter(x, y, c='g')
# labelling graph
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot')
plt.savefig('Test.png')
```

```python
y = x*x
```

```python
# Plot Plot
plt.plot(x,y, 'r*', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D diagram')
```

```python
# Creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo--')
```

```python
# Another Example
x = np.arange(1,11)
y = 3 * x + 5
plt.title("Matplotlib demo")
plt.xlabel('x axis demo')
plt.ylabel()
```

```python
np.pi
```

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title("sine wave form")

# Plot the points using matplotlib
plt.plot(x, y)
```

```python
# Subplot()
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin, 'g--')
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x, y_cos, 'r--')
plt.title('Cosine')
```

```python
# Barplot
x = [2,8,10]
y = [11, 16, 9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x, y)
plt.bar(x2, y2, color='g')
```

```python
# Histogram
a = np.array([22, 87, 5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
```

```python
# Box plot
data = [np.random.normal(0,std,100) for std in range(1,4)]

# rectangular box plot
plt.boxplot(data, vert=True, patch_artist=True);
```

```python
# Pie Chart
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)

plt.axis('equal');
```
